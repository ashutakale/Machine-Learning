##### Predict canada's per capita income in year 2020. There is an exercise folder here on github at same level as this notebook, download that and you will find canada_per_capita_income.csv file. Using this build a regression model and predict the per capita income fo canadian citizens in year 2020

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline
```

```python
df = pd.read_csv('downloads/ML/canada_per_capita_income.csv')
df
```

```python
plt.xlabel('year')
plt.ylabel('capitaincome')
plt.scatter(df.year, df.capitaincome, marker = '+', color = 'red')
```

```python
x = df.drop('capitaincome', axis = 'columns')
x
```

```python
y = df.capitaincome
y
```

```python
reg = linear_model.LinearRegression()
reg.fit(x, y)
```

```python
# Capital income in year 2024
reg.predict([[2024]])
```

```python
# Capital income in year 2020
reg.predict([[2020]])
```

```python
# Accuracy of the model
reg.score(x,y)
```
